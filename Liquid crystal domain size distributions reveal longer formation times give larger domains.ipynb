```python
# Import Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load my data from ImageJ csv output
customt = pd.read_csv('CustomT.csv')
```

# Method

Nanotube liquid crystal films were vacuum-deposited, then transferred to glass slides. All the Variable Volume samples contain 20 µg nanotubes diluted in the listed volume. Cross polarized light microscope images of the samples were captured with a Spot B/W camera (16-bit) and 40x objective.

Image files were loaded into ImageJ where the following processing was applied:  
1. Contrast enhancement, 0.4% of pixels saturated  
2. 3x3 pixel minimum filter  
3. Sharpen  
4. Manually threshold image to mask dark domain spots  
5. Converto to binary  
6. Watershed separation  
7. Count Particles  

The following are histograms of the particle areas counted in each sample. Graph scales are all the same, however the bin sizes vary since some samples span a much larger area range than others. Areas are in cm².
